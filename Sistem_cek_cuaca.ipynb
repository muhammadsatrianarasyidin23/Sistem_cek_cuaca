```python
from datetime import datetime
import datetime
import os
import sys

suhu_template= {
    "tempat": "",
    "suhu": 10,
    "rekomendasi": ""
}

data_suhu={}

def hasil_data():
    print(f"{"Tanggal":<12} {"Lokasi":<19} {"Suhu":<21} {"Rekomendasi dan Cuaca":<40}")
    for tanggal, cuaca in sorted(data_suhu.items(),key=lambda x: datetime.datetime.strptime(x[0], "%d/%m/%Y")):
      print(f"{tanggal:<12} {cuaca["tempat"]:<20} {cuaca["suhu"]:<6} {cuaca["rekomendasi"]:<30}")

def tambah_data():
  while True:
    os.system=("cls")
    cuaca= dict.fromkeys(suhu_template.keys())
    TANGGAL= int(input("Masukkan tanggal="))
    BULAN= int(input("Masukkan bulan="))
    TAHUN= int(input("Masukkan tahun="))
    tanggal= datetime.date(TAHUN, BULAN, TANGGAL).strftime("%d/%m/%Y")
    if tanggal in data_suhu:
        print("Data sudah tersimpan")
        print("Silahkan masukkan data lagi")
        continue
    cuaca["tempat"]= input("Masukkan Lokasi=")
    cuaca["suhu"]= int(input("Masukkan suhu="))
    data_suhu.update({tanggal:cuaca})
    if cuaca["suhu"] < 15:
      cuaca["rekomendasi"]=(f"{"Cuaca Dingin\n":<53} {"Rekomendasi tips kesehatan dan pakaian:\n":<80} {"Gunakan selimut tebal dan gunakan pakaian jaket atau hoodie"}")
    elif  cuaca["suhu"] >=15 and  cuaca["suhu"] <=25:
      cuaca["rekomendasi"]=(f"{"Cuaca Sejuk\n":<52} {"Rekomendasi tips kesehatan dan pakaian:\n":<80} {"Aktivitas luar ruangan dan gunakan  fleece atau sweeter rajut"}")
    elif  cuaca["suhu"] >=26 and  cuaca["suhu"] <=32:
      cuaca["rekomendasi"]=(f"{"Cuaca Hangat\n":<53} {"Rekomendasi tips kesehatan dan pakaian:\n":<81}{"Berolahraga dan gunakan baju longgar bahan katun"}")
    elif  cuaca["suhu"] >=33 and  cuaca["suhu"] <=36:
      cuaca["rekomendasi"]=(f"{"Cuaca Panas\n":<52} {"Rekomendasi tips kesehatan dan pakaian\n":<79} {"Perbanyak minum dan gunakan sunscreen saat keluar rumah serta gunakan pakaian tipis"}")
    elif  cuaca["suhu"] >36:
      cuaca["rekomendasi"]=(f"{"Cuaca terlalu Panas\n":<60} {"Rekomendasi tips kesehatan dan pakaian\n":<79} {"Disarankan jangan keluar rumah jika tidak mau terbakar serta gunakan pakaian tipis dan berwarna"}")
    hasil_data()
    while True:
      kembali= input("Apakah anda ingin tambah daftar suhu lagi?(y/n)")
      if kembali=="N" or kembali=="n":
        menu()
      elif kembali=="Y" or kembali=="y":
        tambah_data()
      else:
        print("Input yang anda masukkan tidak sesuai")


def suhu_hari_ini():
  while True:
    os.system= ("cls")
    print("\n")
    suhu = int(input("Masukkan suhu=(°C)"))
    tempat= input("Masukkan lokasi=")
    tanggal= datetime.datetime.now()
    hari_ini= tanggal.strftime("%d/%m/%Y")
    cuaca= {
        "tempat":tempat,
        "suhu":suhu,
        "rekomendasi":""
    }
    print("waktu:", hari_ini)
    if suhu< 15:
        cuaca["rekomendasi"]=(f"{"Cuaca Dingin\n":<53} {"Rekomendasi tips kesehatan dan pakaian:\n":<80} {"Gunakan selimut tebal dan gunakan pakaian jaket atau hoodie"}")
        print("Cuaca Dingin\n"
        "Rekomendasi tips kesehatan dan pakaian:\n"
        "Gunakan selimut tebal dan gunakan pakaian jaket atau hoodie\n")
    elif suhu >=15 and suhu <=25:
        cuaca["rekomendasi"]=(f"{"Cuaca Sejuk\n":<52} {"Rekomendasi tips kesehatan dan pakaian:\n":<80} {"Aktivitas luar ruangan dan gunakan  fleece atau sweeter rajut"}")
        print("Cuaca Sejuk\n"
        "Rekomendasi tips kesehatan dan pakaian:\n"
        "Aktivitas luar ruangan dan gunakan  fleece atau sweeter rajut")
    elif suhu >=26 and suhu <=32:
        cuaca["rekomendasi"]=(f"{"Cuaca Hangat\n":<53} {"Rekomendasi tips kesehatan dan pakaian:\n":<81}{"Berolahraga dan gunakan baju longgar bahan katun"}")
        print("Cuaca Hangat\n"
        "Rekomendasi tips kesehatan dan pakaian:\n"
        "Berolahraga dan gunakan baju longgar bahan katun")
    elif suhu >=33 and suhu <=36:
        cuaca["rekomendasi"]=(f"{"Cuaca Panas\n":<52} {"Rekomendasi tips kesehatan dan pakaian\n":<79} {"Perbanyak minum dan gunakan sunscreen saat keluar rumah serta gunakan pakaian tipis"}")
        print("Cuaca Panas\n"
        "Rekomendasi tips kesehatan dan pakaian:\n"
        "Perbanyak minum dan gunakan sunscreen saat keluar rumah serta gunakan pakaian tipis")
    elif suhu >36:
        cuaca["rekomendasi"]=(f"{"Cuaca terlalu Panas\n":<60} {"Rekomendasi tips kesehatan dan pakaian\n":<79} {"Disarankan jangan keluar rumah jika tidak mau terbakar serta gunakan pakaian tipis dan berwarna"}")
        print("Cuaca terlalu Panas\n"
        "Rekomendasi tips kesehatan dan pakaian:\n"
        "Disarankan jangan keluar rumah jika tidak mau terbakar serta gunakan pakaian tipis dan berwarna")
    data_suhu[hari_ini]=cuaca
    while True:
      tambah= input("Apakah anda ingin mengecek suhu lagi?(y/n)")
      if tambah=="N" or tambah=="n":
        menu()
      elif tambah=="Y" or tambah=="y":
        suhu_hari_ini()
      else:
        print("Input yang anda masukkan salah")

def cari_data():
  hasil_data()
  while True:
    os.system= ("cls")
    lihat= input("Apakah anda ingin mencari data?(y/n)")
    if lihat=="N" or lihat=="n":
      keluar= input("Apakah anda ingin kembali ke menu utama?(y/n)")
      if keluar=="Y" or keluar=="y":
        menu()
      elif keluar=="N" or keluar=="n":
        continue
      else:
        print("Input yang anda masukkan salah")
        continue
    elif lihat=="Y" or lihat=="y":
      cari= input("Masukkan data apa yang ingin anda cari?(TANGGAL/LOKASI/SUHU)").upper()
      if cari=="LOKASI":
        lokasi= input("Masukkan lokasi yang anda cari=")
        for cuaca in data_suhu:
          if lokasi in data_suhu[cuaca]["tempat"]:
            key=cuaca
            tempat= data_suhu[key]["tempat"]
            suhu= data_suhu[key]["suhu"]
            rekomendasi= data_suhu[key]["rekomendasi"]
            print(f"{"Tanggal":<12} {"Lokasi":<19} {"Suhu":<21} {"Rekomendasi dan Cuaca":<40}")
            print(f"{key:<12} {tempat:<20} {suhu:<6} {rekomendasi:<30}")
          elif lokasi not in data_suhu[cuaca]["tempat"]:
            print("Data tidak ditemukan")
      elif cari=="SUHU":
        suhu= int(input("Masukkan suhu yang anda cari="))
        for cuaca in data_suhu:
          if suhu == data_suhu[cuaca]["suhu"]:
            key=cuaca
            tempat= data_suhu[key]["tempat"]
            suhu= data_suhu[key]["suhu"]
            rekomendasi= data_suhu[key]["rekomendasi"]
            print(f"{"Tanggal":<12} {"Lokasi":<19} {"Suhu":<21} {"Rekomendasi dan Cuaca":<40}")
            print(f"{key:<12} {tempat:<20} {suhu:<6} {rekomendasi:<30}")
          elif suhu is not data_suhu[cuaca]["suhu"]:
            print("Data tidak ditemukan")
      elif cari=="TANGGAL":
        TANGGAL= int(input("Masukkan tanggal yang anda cari="))
        BULAN= int(input("Masukkan bulan yang anda cari="))
        TAHUN= int(input("Masukkan tahun yang anda cari="))
        tanggal= datetime.date(TAHUN, BULAN, TANGGAL).strftime("%d/%m/%Y")
        for cuaca in data_suhu:
          if tanggal in data_suhu:
            key=cuaca
            tempat= data_suhu[key]["tempat"]
            suhu= data_suhu[key]["suhu"]
            rekomendasi= data_suhu[key]["rekomendasi"]
            print(f"{"Tanggal":<12} {"Lokasi":<19} {"Suhu":<21} {"Rekomendasi dan Cuaca":<40}")
            print(f"{key:<12} {tempat:<20} {suhu:<6} {rekomendasi:<30}")
          elif tanggal not in data_suhu:
            print("Data tidak ditemukan")
      else:
        print("Input yang anda masukkan tidak tersedia")
    else:
      print("Input yang anda masukkan tidak tersedia")
    while True:
      tambah= input("Apakah anda ingin kembali ke menu utama?(y/n)")
      if tambah=="N"or tambah=="n":
        break
        continue
      elif tambah=="Y"or tambah=="y":
        menu()
      else:
        print("Input yang anda masukkan salah")

def hapus_data():
  hasil_data()
  while True:
    os.system=("cls")
    TANGGAL= int(input("Masukkan tanggal yang ingin dihapus="))
    BULAN= int(input("Masukkan bulan yang ingin dihapus="))
    TAHUN= int(input("Masukkan tahun yang ingin dihapus="))
    hapus=datetime.date(TAHUN, BULAN, TANGGAL).strftime("%d/%m/%Y")
    if hapus in data_suhu:
      del data_suhu[hapus]
      hasil_data()
    elif hapus not in data_suhu:
      print("Data tidak ditemukan")
    while True:
      tambah= input("Apakah anda ingin hapus data lagi?(y/n)")
      if tambah=="N"or tambah=="n":
        menu()
      elif tambah=="Y"or tambah=="y":
        break
        continue
      else:
        print("Input yang anda masukkan salah")

def ubah_data():
  hasil_data()
  while True:
    os.system=("cls")
    TANGGAL= int(input("Masukkan tanggal yang ingin diubah="))
    BULAN= int(input("Masukkan bulan yang ingin diubah="))
    TAHUN= int(input("Masukkan tahun yang ingin diubah="))
    tanggal_lama= datetime.date(TAHUN, BULAN, TANGGAL).strftime("%d/%m/%Y")
    if tanggal_lama not in data_suhu:
      print("Data tidak ditemukan")
      while True:
        ubah= input("Apakah anda ingin mengubah data?(y/n)")
        if ubah=="N" or ubah=="n":
          menu()
        elif ubah=="Y" or ubah=="y":
          ubah_data()
        else:
          print("Input yang anda masukkan salah")
    elif tanggal_lama in data_suhu:
      Tanggal= int(input("Masukkan tanggal="))
      Bulan= int(input("Masukkan bulan="))
      Tahun= int(input("Masukkan tahun="))
      tanggal_baru= datetime.date(Tahun, Bulan, Tanggal).strftime("%d/%m/%Y")
      suhu= int(input("Masukkan suhu=(°C)"))
      tempat= input("Masukkan lokasi=")
      cuaca= {
      "tempat":tempat,
      "suhu":suhu,
      "rekomendasi":""
      }
      if cuaca["suhu"] < 15:
        cuaca["rekomendasi"]=(f"{"Cuaca Dingin\n":<53} {"Rekomendasi tips kesehatan dan pakaian:\n":<80} {"Gunakan selimut tebal dan gunakan pakaian jaket atau hoodie"}")
      elif  cuaca["suhu"] >=15 and  cuaca["suhu"] <=25:
        cuaca["rekomendasi"]=(f"{"Cuaca Sejuk\n":<52} {"Rekomendasi tips kesehatan dan pakaian:\n":<80} {"Aktivitas luar ruangan dan gunakan  fleece atau sweeter rajut"}")
      elif  cuaca["suhu"] >=26 and  cuaca["suhu"] <=32:
        cuaca["rekomendasi"]=(f"{"Cuaca Hangat\n":<53} {"Rekomendasi tips kesehatan dan pakaian:\n":<81}{"Berolahraga dan gunakan baju longgar bahan katun"}")
      elif  cuaca["suhu"] >=33 and  cuaca["suhu"] <=36:
        cuaca["rekomendasi"]=(f"{"Cuaca Panas\n":<52} {"Rekomendasi tips kesehatan dan pakaian\n":<79} {"Perbanyak minum dan gunakan sunscreen saat keluar rumah serta gunakan pakaian tipis"}")
      elif  cuaca["suhu"] >36:
        cuaca["rekomendasi"]=(f"{"Cuaca terlalu Panas\n":<60} {"Rekomendasi tips kesehatan dan pakaian\n":<79} {"Disarankan jangan keluar rumah jika tidak mau terbakar serta gunakan pakaian tipis dan berwarna"}")
      del data_suhu[tanggal_lama]
      data_suhu[tanggal_baru]=cuaca
      hasil_data()
      while True:
        tambah= input("Apakah anda ingin mengubah data lagi?(y/n)")
        if tambah=="N"or tambah=="n":
          menu()
        elif tambah=="Y"or tambah=="y":
          break
          continue
        else:
          print("Input yang anda masukkan salah")

def menu():
  print("SELAMAT DATANG DI SISTEM CEK CUACA BERDASARKAN SUHU")
  print("="*50)
  print("1. Cek cuaca hari ini")
  print("2. Tambah daftar cuaca")
  print("3. Lihat dan cari daftar cuaca")
  print("4. Hapus daftar cuaca")
  print("5. Ubah daftar cuaca")
  print("6. Keluar")
  pilih= int(input("Masukkan pilihan anda="))
  if pilih==1:
    suhu_hari_ini()
  elif pilih==2:
    tambah_data()
  elif pilih==3:
    cari_data()
  elif pilih==4:
    hapus_data()
  elif pilih==5:
    ubah_data()
  elif pilih==6:
    print("Program selesai")
    print("Terima Kasih telah menggunakan program kami")
    sys.exit()
  else:
    print("Input yang anda masukkan salah")
    menu()

menu()
```
